```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import transformers
import torch

# load model
model = "tiiuae/falcon-7b-instruct"
tokenizer = AutoTokenizer.from_pretrained(model)

falcon_pipeline = transformers.pipeline("text-generation",
                                        model=model,
                                        tokenizer=tokenizer,
                                        torch_dtype=torch.bfloat16,
                                        trust_remote_code=True,
                                        device_map="auto"
                                        )
```

## Configuración

```python
# define completion function
def get_completion_falcon(input):
  # prompt = f"You are an expert Physicist. Your job is to explain complex Physics concepts in simple words. \n{input}"
  # print(prompt)
  system = f"""
  You are an expert Physicist.
  You are good at explaining Physics concepts in simple words.
  Help as much as you can.
  """
  prompt = f"#### System: {system}\n#### User: \n{input}\n\n#### Response from falcon-7b-instruct:"
  print(prompt)
  falcon_response = falcon_pipeline(prompt,
                                    max_length=500,
                                    do_sample=True,
                                    top_k=10,
                                    num_return_sequences=1,
                                    eos_token_id=tokenizer.eos_token_id,
                                    )

  return falcon_response
```

Retomando la Tabla "df" pero en inglés

```python
import pandas as pd

data = {'Topic': ['Classical Mechanics', 'Thermodynamics', 'Electromagnetism', 'Optics', 'Modern Physics', 'Biophysics'], 'Subtopics': ['Kinematics, Dynamics, Rotational Motion, Oscillations and Waves, Gravitation', 'Temperature and Heat, First Law, Second Law, Third Law', 'Electrostatics, Electric Current, Magnetism, Electromagnetic Induction, Electromagnetic Waves', 'Geometric Optics, Physical Optics', 'Special Relativity, Quantum Physics, Nuclear Physics', 'Fluid Physics, Solid State Physics, Acoustics, Astrophysics']}
df = pd.DataFrame(data)
df
```

## Consulta #1

```python
prompt = "In one word... Who is the main scientist who studied Classical Mechanics"
response = get_completion_falcon(prompt)
print(response[0]['generated_text'])
```

## Pregunta en múltiples celdas

```python
prompt_0 = "In one word... Who is the main scientist who studied: "
df['Representante'] = None

for i in range(6):
  prompt = prompt_0 + df["Topic"][i]
  response = get_completion_falcon(prompt)
  df['Representante'][i] = response[0]['generated_text']
```

### A un texto en particular

```python
import re

# Texto con el ejemplo de Isaac Newton
texto_1 = "The main scientist who studied classical mechanics is called Isaac Newton."

# Texto con el ejemplo de James Watson
texto_2 = """#### System:
  You are an expert Physicist.
  You are good at explaining Physics concepts in simple words.
  Help as much as you can.

#### User:
In one word... Who is the main scientist who studied: Biophysics

#### Response from falcon-7b-instruct:
The main scientist who studied biophysics is James Watson, who discovered the structure of DNA."""

# Expresión regular para extraer el nombre del científico
regex = r"(?:called|is) (\b[A-Z][a-z]*\b \b[A-Z][a-z]*\b)"

# Buscar el nombre en los textos
nombre_1 = re.search(regex, texto_1)
nombre_2 = re.search(regex, texto_2)

if nombre_1:
    print(nombre_1.group(1))  # Debería imprimir "Isaac Newton"
if nombre_2:
    print(nombre_2.group(1))  # Debería imprimir "James Watson"
```

### A un DataFrame

```python
import pandas as pd
import re

def extract_scientist_name(text):
    regex = r"(?:called|is|was) (\b[A-Z][a-z]*\b \b[A-Z][a-z]*\b|\b[A-Z][a-z]*\b)"
    #regex = r"(?:called|is) (\b[A-Z][a-z]*\b \b[A-Z][a-z]*\b)"
    match = re.search(regex, text)
    return match.group(1) if match else None

# Aplicar la función a la columna 'Representante'
df['scientist_name'] = df['Representante'].apply(extract_scientist_name)
```
